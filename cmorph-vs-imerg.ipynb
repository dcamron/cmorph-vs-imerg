## Imports and Setup

```python
import warnings
# imported just to hide xarray import warning
warnings.simplefilter('ignore')
import pandas as pd
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# CMORPH
## Initial January 2016 Monthly Mean

```python
ds = xr.open_mfdataset('data/cmorph/2016/01/*.cdf',
                       autoclose=True, concat_dim='time')
```

```python
monmean = ds.mean(dim='time').Precipitation
```

```python
fig = plt.figure(figsize=(18, 9))
fig.suptitle('CMORPH Monthly Mean Precipitation January 2016')
ax = plt.axes(projection=ccrs.Robinson(central_longitude=180))
c = monmean.plot.contourf(
    ax=ax, transform=ccrs.PlateCarree(), cmap='Greens', add_colorbar=False)
cb = plt.colorbar(c,ax=ax)
cb.set_label('[mm/hr]')
ax.coastlines()
ax.gridlines()
```
